```python
import random 
import numpy as np
#import pandas as pd
import matplotlib.pyplot as plt
```

```python
n=range(1,500)
```

```python
N=[]
v=[]
for j in n:
    Sum=0
    for i in range(j):
        Sum+=random.choice([1,0])
    N.append(Sum)
    v.append(Sum/j)
```

```python
plt.plot(n,v)
plt.xlabel("n")
plt.ylabel("V = avg/n")
```

```python
coins=1000
tosses=10000
```

```python
listsums=[]
for i in range(tosses):
    ching=random.choices([1,0],k=coins)
    listsums.append(sum(ching))
```

```python

plt.figure()
plt.xlabel("Sum of {0} Coin flips".format(coins))
plt.ylabel("corresponding Frequncy; for {0} tosses".format(tosses))

plt.hist(listsums,bins='auto', color='b', alpha=0.9, rwidth=0.75,density=True)
```
