```python
import numpy as np
import sympy as sp
from sympy.physics.hydrogen import R_nl
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
R,n,l,Z,a,p,L,k=sp.symbols('R n l Z a p L k')
```

```python
p=2*Z/(n*a)
```

```python
L=sp.Sum( pow(-1,k+1) * pow(sp.factorial(n+l),2) *pow(p,k) /( sp.factorial(n-l-1-k)*sp.factorial(2*l+1+k)*sp.factorial(k) )  ,(k,0,n-l-1))
```

```python
R = pow( (-2*Z/(n*a)) ,3/2) *pow( sp.factorial(n-l-1) / (2*n*pow(sp.factorial(n+l),3) ) ,1/2) * sp.exp( -p/2 ) *pow(p,l)*L
```
