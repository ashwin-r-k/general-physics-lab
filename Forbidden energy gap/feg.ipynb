```python
%reset -f
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
import pandas as pd
import statistics
from scipy.stats import norm
```

```python
df = pd.read_excel(r'/home/ashwin/Git/general-physics-lab/Forbidden energy gap/feg.xlsx')
print(df)
```

```python
T=df['T'].to_numpy()
V=df['V'].to_numpy()
```

```python
x=V[:]
y=T[:]
deg=1

result=np.polyfit(x,y,deg)


p = np.poly1d(result)

xp = np.linspace(min(x),max(x))

plt.figure()
plt.plot(x, y, 'og',label="Observed Data Points")
plt.plot(xp, p(xp),'k',label="Curve fitting as a polynomial")
plt.xlabel("V")
plt.ylabel("T")
plt.grid("TRUE")

plt.legend()

```

```python
x=V[:]
y=T[:]
result = linregress(V, T)

fig = plt.figure()
ax = fig.add_subplot(111)

m = result.slope
b = result.intercept
delta_m = result.stderr 
#delta_b = result.intercept_stderr 

plt.scatter(V,T)
plt.plot(V, np.array(V) * m + b , color="red")  

ax.scatter(V, T,label = 'observation')

ax.plot(V, np.array(V) * m + b , color="red",label = 'least squares regression')
ax.plot(V, np.array(V) * (m +delta_m ) + b , color="green",label = 'standard deviation')
ax.plot(V, np.array(V) * (m-delta_m) + b , color="green")

plt.xlabel(r"V")
plt.ylabel(r"T")
plt.grid("TRUE")
plt.title("curve: T = [{0}$\pm{1}$ ] V + {2} ".format(round(m,10),round(delta_m,10),round(b,2) ))
ax.legend()
plt.show()
```
